## What are Vector Stores??

### **Vector stores are essentially databases that are designed for a specific kind of data format — vector embeddings. They specialize in storing these embeddings very efficiently. Vector stores enable fast retrieval based on similarity searches.**

### **Essentially the unstructured data is loaded, transformed and embedded into vectors and stored in a vector store. During query time, the unstructured query is embedded and the most similar embedding vectors with respect to the embedded query is retrieved. Vector stores takes care of storing the embedded data and performing vector search.**

## Import all necessary dependencies

```python
from langchain_community.document_loaders import WikipediaLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
```

## Step1: Document Loading- Let's use WikipediaLoader as the DocumentLoader.

```python
!pip install wikipedia
```

```python
loader = WikipediaLoader(query='Artificial Intelligence',load_max_docs=4)
documents = loader.load()
print(documents)
```

```python
print(len(documents))
```

```python
for i,document in enumerate(documents):
  print()
  print(f'***************** document: {i} ******************')
  print()
  print(document.page_content)
  print()
```

## Step2: Text Splitting - We will use `RecursiveCharacterTextSplitter` as the text splitter.

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=400,chunk_overlap=100)
docs = text_splitter.split_documents(documents)
print(docs)
```

```python
print(len(docs))
## means total 51 splitted docs
```

```python
## first docs
print(docs[0].page_content)
```

```python
## last docs
print(docs[50].page_content)
```

## Step3: Embedding Function - For the embedding function, we will use HuggingFace BGE Embedding model and let's create a query.

```python
from langchain.embeddings import HuggingFaceBgeEmbeddings

model_name = "BAAI/bge-large-en-v1.5"
model_kwargs={'device':'cpu'}
encode_kwargs={'normalize_embeddings':True}

embeding_function=HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)

query = 'who is father of ms dhoni?'
```

# FAISS

### Let’s look into the functionalities of FAISS vector database, which makes use of the Facebook AI Similarity Search (FAISS) library.

```python
!pip install faiss-cpu
```

```python
from langchain_community.vectorstores import FAISS

db = FAISS.from_documents(
    docs,
    embedding=embeding_function,
)
```

### Next, we can query the vector database. We can either do similarity search or similarity search with vector.

```python
matched_docs = db.similarity_search(query="what is generative ai?",k=2)
matched_docs
```

## we can also emdding query passed to db.

```python
embedding_vector = embeding_function.embed_query("what is generative ai?")
matched_docs = db.similarity_search_by_vector(embedding_vector,k=2)
matched_docs
```

### Most of the times, we won’t necessarily use an in-memory database. Let’s create another open-source database.

## ChromaDB

### ChromaDB can be used as both an in-memory database and a backend with persistent storage options.

```python
!pip install chromadb
```

```python
from langchain.vectorstores import Chroma
db = Chroma.from_documents(
    docs,
    embeding_function,
    persist_directory='output/ai_db'
)
```

## Loading the database

```python
loaded_db = Chroma(persist_directory='output/ai_db',embedding_function=embeding_function)
```

### Similarly, here also we can do similarity search as well as similarity search with vector.

```python
matched_docs = loaded_db.similarity_search(query='what is gen ai?')
matched_docs
```

```python
## also query embedding
embedding_vector = embeding_function.embed_query("what is generative ai?")
matched_docs = loaded_db.similarity_search_by_vector(embedding_vector,k=2)
matched_docs
```
